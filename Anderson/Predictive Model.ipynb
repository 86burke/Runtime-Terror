```python
import pandas as pd
import numpy as np
```

```python
qbdf = pd.read_csv("QBStats.csv")
qbdf19 = pd.read_csv("QBStats2019.csv")
```

```python
QBMeasures = qbdf[["Player Name","Interceptions","Passing Yards", "Passing TD","Fumbles","Passing Completions","Passing Attempts"]]
QBMeasures
```

```python
QBMeasures19 = qbdf19[["Player Name","Interceptions","Passing Yards", "Passing TD","Fumbles","Passing Completions","Passing Attempts"]]
QBMeasures19
```

```python
PassingYPTD = QBMeasures['Passing Yards']/QBMeasures['Passing TD']
PassingCPA = QBMeasures['Passing Completions']/QBMeasures['Passing Attempts']
PassingYPTD19 = QBMeasures19['Passing Yards']/QBMeasures19['Passing TD']
PassingCPA19 = QBMeasures19['Passing Completions']/QBMeasures19['Passing Attempts']
```

```python
QBMeasures["Passing Yards Per TD"] = PassingYPTD
QBMeasures19["Passing Yards Per TD"] = PassingYPTD19
```

```python
QBMeasures["Passing Completions Per Attempts"] = PassingCPA
QBMeasures19["Passing Completions Per Attempts"] = PassingCPA19
```

```python
QBMeasures19
```

```python
QBMeasuresFinal = QBMeasures[['Player Name','Interceptions','Passing Yards Per TD','Fumbles','Passing Completions Per Attempts']]
QBMeasuresFinal19 = QBMeasures19[['Player Name','Interceptions','Passing Yards Per TD','Fumbles','Passing Completions Per Attempts']]
```

```python
QBMeasuresFinal.head
QBMeasuresFinal = QBMeasuresFinal.set_index('Player Name')
QBMeasuresFinal19 = QBMeasuresFinal19.set_index('Player Name')
```

```python
QBMeasuresFinal
QBMeasuresFinal.replace([np.inf, -np.inf], np.nan)
```

```python
QBMeasuresFinal19
QBMeasuresFinal19.replace([np.inf, -np.inf], np.nan)
```

```python
QBMeasuresFinal = QBMeasuresFinal.replace([np.inf, -np.inf], 0)
QBMeasuresFinal19 = QBMeasuresFinal19.replace([np.inf, -np.inf], 0)
```

```python
QBMeasuresFinal = QBMeasuresFinal.fillna(0)
QBMeasuresFinal19 = QBMeasuresFinal19.fillna(0)
```
